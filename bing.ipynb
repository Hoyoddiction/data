```python
def Rating(AvgRating):
    if AvgRating == "NaN":
        return 'NaN'
    elif AvgRating < 5:
        return 'Low'
    elif AvgRating < 6:
        return 'Mid'
    elif AvgRating < 8:
        return 'Up'
    else:
        return 'High'
Mdf['Rating'] = Mdf['AvgRating'].apply(Rating)
Mdf
     
```

```python
def root_element(x):
  if isinstance(x, (int, float)):
    return x ** (1/5)
  else:
    return x

rdf = Mdf.applymap(root_element)
rdf
```
